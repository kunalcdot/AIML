```python
df = pd.read_csv('indian_food.csv')
df
```

```python
df1 = df[['name','ingredients']]
df1
```

```python
df2 = df.head(10)
df2
```

1st plotting method using matplotlib.


```python
x = df2['name']
y = df2['prep_time']
y1 = df2['cook_time']

width = 0.4
cntr_pt = np.arange(len(x))
plt.bar(cntr_pt,y,label='Prep',width = width)
plt.bar(cntr_pt+width,y1,label='Cook',width = width)
plt.xlabel('Name')
plt.xticks(cntr_pt, x) 
plt.xticks(rotation=45)

plt.show()
```
