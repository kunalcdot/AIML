```python
import pandas as pd
import numpy as np
from sklearn import linear_model
# importing matplotlib module 
from matplotlib import pyplot as plt 

file = 'hiring.csv'
df = pd.read_csv(file)
df
df.columns
```

```python
# 
df['experience']
df

```

```python
# 
med_test_score = df['test_score(out of 10)'].mean()
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(med_test_score)
df
```

```python
from word2number import w2n
df['experience'] = df['experience'].fillna('zero')
df['exp in no.'] = df['experience'].apply(w2n.word_to_num)
df
```

```python
features = ['test_score(out of 10)','interview_score(out of 10)','exp in no.']
y = df['salary($)'] # target dependent variable
X = df[features]
model = linear_model.LinearRegression()

model.fit(X,y)
model.predict([[9,6,2],[10,10,12]])
```
