```python
#Plot bar charts showing impact of employee salaries on retention
#Plot bar charts showing corelation between department and employee retention

from matplotlib import pyplot as plt 
x = (df['salary'].unique())

df_low_salary = df[(df['salary'] =='high')]


low_salary_emp = len(df[(df['salary'] =='low')])
mid_salary_emp = len(df[(df['salary'] =='medium')])
high_salary_emp = len(df[(df['salary'] =='high')])
y = [low_salary_emp,mid_salary_emp,high_salary_emp]
#plt.bar(x,y)
#plt.show()
```

```python
low_salary_emp_left = len(df[(df['salary'] =='low')&(df['left'] ==1)])
mid_salary_emp_left = len(df[(df['salary'] =='medium')&(df['left'] ==1)])
high_salary_emp_left = len(df[(df['salary'] =='high')&(df['left'] ==1)])

low_left_perct = (1-(low_salary_emp_left/low_salary_emp))*100
mid_left_perct = (1-(mid_salary_emp_left/mid_salary_emp))*100
high_left_perct = (1-(high_salary_emp_left/high_salary_emp))*100

y = [low_left_perct,mid_left_perct,high_left_perct]
plt.xlabel('Salary')
plt.ylabel('Retention%')
plt.bar(x,y)
plt.show()
```

```python
# department vs retention
#['sales' 'accounting' 'hr' 'technical' 'support' 'management' 'IT''product_mng' 'marketing' 'RandD']
ret_arr = []
dept_arr = df['Department'].unique()
for dept in dept_arr:
    #print(dept)
    ret_val = (len(df[(df['Department'] ==dept)&(df['left'] ==1)]))/(len(df[(df['Department'] ==dept)]))*100
    #print(ret_val)
    ret_arr.append(ret_val)
plt.xlabel('Department')
plt.ylabel('Retention%')
plt.bar(dept_arr,ret_arr)
plt.show()
```

```python
# time_spend_company vs retention

ret_arr = []
time_arr = df['time_spend_company'].unique()
for time in time_arr:
    print(time)
    ret_val = (len(df[(df['time_spend_company'] ==time)&(df['left'] ==1)]))/(len(df[(df['time_spend_company'] ==time)]))*100
    print(ret_val)
    ret_arr.append(ret_val)
plt.xlabel('Time spend')
plt.ylabel('Retention%')
plt.bar(time_arr,ret_arr)
plt.show()
```

```python
df = pd.read_csv('HR_comma_sep.csv')
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['low', 'medium', 'high']])
df['salary_encoded'] = oe.fit_transform(df[['salary']])
df
```

```python
# let's create model using salary and promo
#X = df[['promotion_last_5years','salary_encoded']]
#X = df[['salary_encoded']]
X = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary_encoded']]

Y = df['left']
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X,Y,train_size=0.8)

model = LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
